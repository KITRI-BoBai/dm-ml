## outlier 제외하기<br>
최종 feature를 5개로 할거라 관련 행에 대해서만 이상치 제거함<br>
제거 안한것도 있긴함

###circulation_supply행에서 outlier 없앰



```python
# circulating_supply IQR

quantile_25 = np.quantile(dfdf['circulating_supply'], 0.25)
quantile_75 = np.quantile(dfdf['circulating_supply'],0.75)
IQR = quantile_75 - quantile_25
minimum_circulating_supply = quantile_25 - 1.5*IQR
maximum_circulating_supply = quantile_75 + 1.5*IQR
dfdf2 = dfdf[(minimum_circulating_supply <= dfdf['circulating_supply']) & (dfdf['circulating_supply'] <= maximum_circulating_supply)]
dfdf3 = dfdf[(minimum_circulating_supply > dfdf['circulating_supply']) | (dfdf['circulating_supply'] > maximum_circulating_supply)]
```

```python
#이상치 분포 시각화
fig2 = px.scatter_3d(dfdf3, x="circulating_supply", y="quote.USD.market_cap",z="quote.USD.volume_24h",
                     color="rank",size="rank")
fig2.update_layout(title="5 Features Representation")
fig2.show()
```

```python
#정상치 분포 시각화
fig2 = px.scatter_3d(dfdf2, x="circulating_supply", y="quote.USD.market_cap",z="quote.USD.volume_24h",
                     color="rank",size="rank")
fig2.update_layout(title="5 Features Representation")
fig2.show()
```

##종합+시각화

###outlier제거한 애들의 교집합 구함

```python
#2가 정상치 3은 이상치임
dfdf_total_2 = dfdf[(minimum_volume_24h <= dfdf['quote.USD.volume_24h']) & (dfdf['quote.USD.volume_24h'] <= maximum_volume_24h)
            & (minimum_market_cap <= dfdf['quote.USD.market_cap']) & (dfdf['quote.USD.market_cap'] <= maximum_market_cap)
            & (minimum_circulating_supply <= dfdf['circulating_supply']) & (dfdf['circulating_supply'] <= maximum_circulating_supply)]
dfdf_total_3 = dfdf.mask(
                (minimum_volume_24h <= dfdf['quote.USD.volume_24h']) & (dfdf['quote.USD.volume_24h'] <= maximum_volume_24h)
            & (minimum_market_cap <= dfdf['quote.USD.market_cap']) & (dfdf['quote.USD.market_cap'] <= maximum_market_cap)
            & (minimum_circulating_supply <= dfdf['circulating_supply']) & (dfdf['circulating_supply'] <= maximum_circulating_supply)
)
```

```python
total_df = dfdf_total_2
```

###feature 전처리 <br>
선정&처리 기준은 내 뇌피셜(==신뢰할수없음)<br>
**supply_ratio** : 순환공급량 / 전체 공급량<br>
**inverse_market_cap** : 시가총액 역수에 로그취하고 양수로 만듦(시가총액 단위가 너무 커서)<br>
**inverse_volume** : 거래량 역수에 로그취하고 양수로 만듦<br>
**inverse_num_market_pairs** : 상장된 거래소 수 역수에 로그취하고 양수로 만듦

```python
total_df["supply_ratio"] = total_df["circulating_supply"] / total_df["total_supply"]
total_df["inverse_market_cap"] = np.log(1/(total_df['quote.USD.market_cap']+1))*(-1)
total_df["inverse_volume"] = np.log(1/(total_df['quote.USD.volume_24h']+1))*(-1)
total_df['inverse_num_market_pairs'] = np.log(1/(total_df['num_market_pairs']+1))*(-1)
```

###이게 최종 시각화<br>
x축은 inverse_volume, y축은 supply_ratio, z축은 inverse_num_market_pairs임<br>
색이 진할수록 coinmarketcap에서의 순위가 높음<br>
동그라미 사이즈는 시총 커질수록 크다

```python
#얘는 혼자 튀어서 뺐음
result = total_df[total_df['rank']!=1046]
```

```python
fig2 = px.scatter_3d(result, x="inverse_volume", y="supply_ratio",z="inverse_num_market_pairs",
                     color="rank",size="inverse_market_cap")
fig2.update_layout(title="5 Features Representation")
fig2.show()
```
